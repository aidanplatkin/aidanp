```python
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1,6])
ypoints = np.array([5,10])

plt.plot(xpoints, ypoints)
plt.show
```

```python
xpoints = np.array([3,6,9,12])
ypoints = np.array([2,8,1,10])

plt.plot(xpoints, ypoints, color = 'blue')
plt.plot(xpoints, ypoints, marker = 'o')
plt.show()
```

```python
xpoints = np.array([0,1,2,3,4,5])
ypoints = np.array([2,4,6,14,10,12])

plt.plot(xpoints, ypoints, 'D:r', mfc = 'g', mec ='g')
```

```python
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([9,10,11,12,13,14])

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([7,8,9,10,11,12])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([5,6,7,8,9,10])

plt.plot(x1,y1,x2,y2,x3,y3)
plt.show()
```

```python
students = {"Andy":88,
           "Amy":66,
           "James": 90,
           "Jules": 55,
           "Arthur": 77}
names = list(students.keys())
values = list(students.values())

plt.bar(range(len(students)), values, tick_label=names)
plt.show()
values = list(students.values())

plt.show()
```

```python
students = {"Andy":88,
           "Amy":66,
           "James": 90,
           "Jules": 55,
           "Arthur": 77}
names = list(students.keys())
values = list(students.values())

plt.pie(values, labels = names) 
plt.legend()
plt.show()
```

```python
x = np.array([2, 4, 6, 4, 5 ,7, 8, 1])
y = np.array([7, 3, 9, 1 ,5 ,8, 2, 7])

x1 = np.array([2, 3, 8])
y1 = np.array([7, 5, 4])

plt.subplot(2, 3, 1)
plt.plot(x,y,x1,y1)
plt.title("GROUPS")


plt.subplot(2, 3, 2)
plt.bar(x, y)
plt.title('NUMBERS')


plt.subplot(2, 3, 3)
plt.pie(x, labels = y)
plt.title("quantity")


plt.subplot(2, 3, 4)
plt.title("Cost over Tine")
plt.xlabel("Time")
plt.ylabel("Cost")
plt.plot(x, y)

plt.grid()


plt.subplot(2, 3, 5)
plt.scatter(x,y)
plt.title("Distance")


plt.subplot(2, 3, 6)
plt.hist(x)
plt.title("shoes")
plt.show()


```
